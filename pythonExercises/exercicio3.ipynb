```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = int(input('Informe o primeiro valor: '))
stop = int(input('Informe o ultimo valor: '))
quantity = int(input('Informe a quantidade de elementos: '))

Ax = np.linspace(start, stop, quantity)
print(Ax)
```

```python
oneThirdOfAx = int(sum(Ax)/3)
print(oneThirdOfAx)
```

```python
Ay = np.linspace(start, stop, quantity+oneThirdOfAx)
print(Ay)
```

```python
plt.plot(Ax, label='Elemento do Vetor', color='blue')
plt.plot(Ay, label='Valor do Elemento do Vetor', color='green')

plt.legend()
plt.show()
```
