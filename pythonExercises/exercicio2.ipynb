```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sunspots = pd.read_csv('sunspots.csv')
```

```python
sunspotsOnDate = sunspots.query('Date >= "2000-01-01"')
print(sunspotsOnDate)
```

```python
y = sunspotsOnDate['Monthly Mean Total Sunspot Number']
plt.plot(y, color='red')
        
plt.show()
```
